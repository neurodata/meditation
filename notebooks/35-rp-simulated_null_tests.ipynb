## Dcorr (500 datasets, d=100, 1000 perms)

```python
with open('../data/2sample_tests/simulations/DCORR_gaussian_SIMULATED_datasets=500_dict_perm=1000_dim=100_min_rank-ZG3.pkl', 'rb') as handle:
    data_dict = pickle.load(handle)
```

```python
fig, axes = plt.subplots(len(data_dict.keys()),1,sharex=True, figsize=(4, 4*len(data_dict.keys())))
if len(data_dict.keys()) == 1:
    axes = [axes]
for ax,test in zip(axes, data_dict.keys()):
    n = len(data_dict[test])
    entries, edges, _ = ax.hist(data_dict[test], bins = np.arange(0,1.1,0.1), weights=np.ones(n)/n, color='b')
    
    # calculate bin centers
    bin_centers = 0.5 * (edges[:-1] + edges[1:])
    ax.axhline(y=sum(entries) / len(bin_centers), ls='--', c='#333333')
    ax.errorbar(
        bin_centers, entries,
        yerr = 1.96 * np.sqrt(entries * (1 - entries) / n),
        fmt='.',c='#333333'
    )
    
    ax.set_title(f'i.e. {test}')
    ax.set_xlim(0,1)
    ax.set_yticks([0,0.1])
# for ax in axes[:-1]:
#     ax.set_xticks([])
axes[-1].set_xlabel('pvalues')
axes[-1].set_xticks(np.arange(0,1.1,0.1))
plt.show()
```
