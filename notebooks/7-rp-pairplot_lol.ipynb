```python
from graspy.plot import pairplot
import pandas as pd
import numpy as np

from pathlib import Path
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
## Define paths
basedir = Path('..')
datadir = basedir / 'data'
```

```python
name = 'lolp_state_gcca_raw-True_param-0.9_var'
df = pd.read_csv(datadir / 'interim' / f'{name}.csv', header = None, index_col=0)
#df = pd.read_csv(datadir / 'interim' / 'lol_embed_gcca_raw-True_param-0.9.csv', header = None, index_col=0)
classes = list(df.index)
X = df.values
```

```python
states = ['Resting State','Compassion','Open Monitoring']
traits = ['Novice', 'Expert']
labels_state = [states[(i-1)%3] for i in classes]
labels_trait = [traits[int(i>3)] for i in classes]
bothnames = {5:'Expert Compassion',
             4:'Expert Resting State',
             6:'Expert Open Monitoring',
             2:'Novice Compassion',
             1:'Novice Resting State',
             3:'Novice Open Monitoring'}
labels_both = [bothnames[i] for i in classes]
```

```python
plot = pairplot(X, labels_both,title='LOL Embeddings, colored by trait/state', palette='Set1')
#plt.savefig(basedir / 'reports' / f'{name}_both.png')
```

```python
plot = pairplot(X, labels_trait,title='LOL Embeddings, colored by trait', palette='Set1')
#plt.savefig(basedir / 'reports' / f'{name}_trait.png')
```

```python
plot = pairplot(X, labels_state,title='LOL Embeddings, colored by state', palette='Set1')
#plt.savefig(basedir / 'reports' / f'{name}_state.png')
```
