```python
from graspy.plot import pairplot
import pandas as pd
import numpy as np

from pathlib import Path
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
## Define paths
basedir = Path('..')
datadir = basedir / 'data'
```

```python
name = 'lolp_state_gcca_raw-True_param-0.9_split-half_ranks-3_6'
df = pd.read_csv(datadir / 'interim' / f'{name}.csv', header = None, index_col=0)
classes = list(df.index)
X = df.values

name = 'lolp_state_gcca_raw-True_param-0.9-split-half_fit_ranks-3_6'
df = pd.read_csv(datadir / 'interim' / f'{name}.csv', header = None, index_col=0)
classes2 = list(df.index)
X2 = df.values
```

```python
#title = 'LOL Embeddings with Shuffled Labels'
title='LOL Embedding'
```

```python
states = ['Resting State','Compassion','Open Monitoring']
traits = ['Novice', 'Expert']
labels_state = [states[(i-1)%3] for i in classes]
labels_trait = [traits[int(i>3)] for i in classes]
bothnames = {5:'Expert Compassion',
             4:'Expert Resting State',
             6:'Expert Open Monitoring',
             2:'Novice Compassion',
             1:'Novice Resting State',
             3:'Novice Open Monitoring'}
labels_both = [bothnames[i] for i in classes]
labels_both2 = [bothnames[i] for i in classes2]
```

```python
plot = pairplot(X, labels_both,title=f'{title}, Non fitted data ,colored by trait/state', palette='Set1',height=2)
plt.savefig(basedir / 'reports' / 'lol' / f'{name}_nonfitted_splithalf.png')
```

```python
plot = pairplot(X2, labels_both2,title=f'{title}, Fitted data, colored by trait/state', palette='Set1',height=2)
plt.savefig(basedir / 'reports' / 'lol' / f'{name}_fitted_splithalf.png')
```

```python
plot = pairplot(X, labels_trait,title=f'{title}, colored by trait', palette='Set1')
#plt.savefig(basedir / 'reports' / 'lol' / f'{name}_shuffled_trait.png')
```

```python
plot = pairplot(X, labels_state,title=f'{title}, colored by state', palette='Set1')
#plt.savefig(basedir / 'reports' / 'lol' / f'{name}_shuffled_state.png')
```
